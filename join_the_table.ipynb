```python
import pandas as pd
import numpy as np
import os
```

```python
customers = pd.read_csv("customers.csv")
sales_orders = pd.read_csv("sales_orders.csv")
sales_orders_header = pd.read_csv("sales_orders_header.csv", parse_dates=["Creation_Date", "Release_Date"])
service_map = pd.read_csv("service_map.csv")
classification = pd.read_csv("classification.csv")
business_units = pd.read_csv("business_units.csv")
```

Filter customers.csv with non-zero Item_Position and match them with both Sales_Order and Item_Position in sales_orders.csv, set the unmatched Item_Position in sales_orders to 0

```python
item_customers = customers.loc[customers["Item_Position"]!=0]
for order_id in item_customers["Sales_Order"].unique():
    temp_customers = item_customers.loc[item_customers["Sales_Order"]==order_id]
    sales_orders.loc[(sales_orders["Item_Position"].isin(temp_customers["Item_Position"]))&(sales_orders["Sales_Order"]==order_id), "IP_Matched"] = True
    # customers.loc[(customers["Sales_Order"]==order_id)&(customers)]
item_customers["Sales_Order"].unique()[1]
```

```python
sales_orders.loc[~(sales_orders["IP_Matched"]==True), "Item_Position"] = 0
```

```python
item_sales_orders = sales_orders.loc[sales_orders["Item_Position"]!=0]
```

```python
item_sales_orders = sales_orders.loc[sales_orders["Item_Position"]!=0]

item_customers = customers.loc[customers["Item_Position"]!=0]

for order_id in item_sales_orders["Sales_Order"].unique():
    temp_order = item_sales_orders.loc[item_sales_orders["Sales_Order"]==order_id]
    customers.loc[(customers["Item_Position"].isin(temp_order["Item_Position"]))&(customers["Sales_Order"]==order_id), "IP_Matched"] = True
customers.loc[~(customers["IP_Matched"]==True), "Item_Position"] = 0
```

```python
sales_orders.drop(columns="IP_Matched", inplace=True)
customers.drop(columns="IP_Matched", inplace=True)
```

```python
df = pd.merge(sales_orders, customers, how="left", on=["Sales_Order", "Item_Position"])
df
```

```python
df.dropna(subset=["Customer_ID", "Material_Class"], inplace=True)
```

```python
sales_orders_header = sales_orders_header.drop(columns=["Net_Value"])
df2 = pd.merge(df, sales_orders_header, how="left", on=["Sales_Order"])
df2
```

```python
df3 = pd.merge(df2, business_units, how="left", on=["Cost_Center"])
df3
```

```python
service_map["Service_Type"] = [i for i in range(29)]
df4 = pd.merge(df3, service_map, how="left", left_on=["Material_Class"], right_on=["MATKL_service"])
df4
```

```python
df5 = pd.merge(df4, classification, how="left", left_on=["Customer_ID"], right_on=["Customer_ID"])
df5
```

```python
path = "training_variables"
if not os.path.exists(path):
    os.mkdir(path)
```
